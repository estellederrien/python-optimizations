## Rôle
Exécuter une simulation M/M/S de file d'attente permet de prendre ses dispositions afin ne pas se retrouver avec des files d'attentes interminables, préjudiciables à la clientèle. On la trouve parfois écrite en M/M/C.
Au préalable, on doit avoir effectué des observations afin de calculer les taux d'arrivée et taux de service.
Par exemple, des personnes arrivant à un guichet automatique au taux λ, attendant leur tour dans la rue et retirant de l’argent au taux μ.

Note: 
Il existe aussi le modèle M/D/1 , le D signifiant un service deterministe au lieu de random.

## Les paramètres
Pour lancer une simulation de file d'attente M/M/S , on doit fournir les paramètres.

- <b>1. Le S dans M/M/S</b> :  C'est le nombre de serveurs.
- <b>2. Le taux d'arrivée (lambda)</b> : C'est le taux  d'arrivée des clients qui suit la loi de probabilité de Poisson ( C'est une distribution de Poisson).
     * nombre de client arrivés / unité de temps
- <b>3. Le taux de services (mu)</b> : C'est le taux de service qui suit la loi de probabilité exponentielle ( C'est une "distribution exponentielle").
    * nombre de clients servis / unité de temps

IMPORTANT : Mu et Lambda doivent être réduits à la même unité de temps, avant de faire les calculs.
Si on nous donne une ARRIVEE de 5 clients mesuré toutes les 1/2 heures, on en déduit qu'il y en a 10 par heures.
Et pareil pour le taux de service, si un serveur assure un rendez vous toutes les 1/2 heures, alors on en déduit qu'il y en a 2 par heure, seulement après cela, on lance le script python!

## Les questions résolues par la théorie des files d'attente.

La théorie des files d'attente peut répondre aux questions suivantes :

- Efficacité de chaque serveur (p) ; si p>1 alors pas d'état état stable possible (steady state)
- Probabilité de zéro ou n client dans le système (p_0, p_n) ;
- Le nombre moyen de clients en attente dans la file d'attente (L_q) ;
- Temps moyen passé par les clients dans la file d'attente (W_q) ;
- Temps moyen que les clients passent dans le système (W) ( Queue + Serveur, loi de Little);
- Nombre moyen de clients dans le système (L) ( Queue + Serveur).


Formules :

- Intensité de serveur, intensité de traffic : p = lambda / mu * S ( S = nb de serveurs)
- Nombre moyen de clients dans le système (L) ( Queue + Serveur) : L = p / 1 - p
- Temps moyen que les clients passent dans le système (W) : W = L / lambda

# Les formules pour un M/M/1

Quand il n'y a qu'un seul serveur, les formules sont plus simples.

Calculs tirés de la documentation du cours de Washington university :
http://courses.washington.edu/inde411/QueueingTheoryPart2.pdf
et confirmés par comparaison avec le site revoledu.com

```python
######################
# Paramètres
######################

# Nombre de serveurs
S = 1

# Taux d'arrivée ( Calculée grâce à la loi de Poisson)
# 0.75 client / 5 minutes soit 0.15 client par minute , on mets à la même unité
Lambda = 0.15

# Taux de service  ( Calculée grâce à la loi exponentielle)
# average service time  1/mu = 1.079 minutes par clients, qui est équivalent à 
# mu = 0.927 clients par minutes
mu = 0.927

######################
# Statistiques
######################

# Intensité de serveur, intensité de trafic ( si P < 1  , alors état stable (steady state OK dans ce cas)): 
P = Lambda / mu 

# Probabilité qu'il n'y ait pas de clients dans le système 
P0 = 1 - P 

# Probabilité qu'il y ait 5 clients dans le système
P_ARRONDI = round(P,3)
P5 = P_ARRONDI ** 5 * P0 

# Nombre moyen de clients dans la file d'attente   
Lq_RevolDU = (Lambda ** 2 ) / ( mu *  (Lambda - mu))
Lq_UniversityWashington = (Lambda ** 2 ) / ( mu *  (mu - Lambda)) 

# Nombre moyen de clients dans le système 
L = Lambda / (mu - Lambda) 
L_eduSite = P / (1 - P) 

# Temps total moyen qu'un client passe dans le système (queue + serveur) 
W = 1 / ( mu - Lambda) 

# Temps total moyen qu'un client attends dans la file d'attente (queue)
Wq = Lambda / mu * ( mu - Lambda) 

# Imprimer le reporting
print("Intensité de queue:",round(P,2),"\n")
print("Pourcentage du temps d'utilisation du serveur:",round(P,2)*100,"%\n")
print("Probabilité P(0) de 0 clients dans la file d'attente:",round(P0,2) * 100,"%\n")
print("Probabilité P(5) de 5 clients dans le système:",P5 * 100,"%\n")
print("Longueur moyenne Lq de la file d'attente (Calcul RevolDU):",round(-Lq_RevolDU,4),"Personnes\n")
print("Longueur moyenne Lq de la file d'attente ( Calcul University Washigton) :",round(Lq_UniversityWashington,4),"Personnes\n")
print("Nombre L moyen de clients dans le système ( University Washington ) :",round(L,4),"Personnes\n")
print("Nombre L moyen de clients dans le système ( Calcul du site EDU ) :",round(L_eduSite,4),"Personnes\n")
print("Temps total W moyen qu'un client passe dans le système( queue + serveur) :",round(W,3)," Minutes\n")
print("Temps total Wq moyen qu'un client attends dans la file d'attente:",round(Wq,3)," Minutes \n")
```

# Un exemple avec une file d'attente M/M/2

Si on a plus de 1 serveurs, les formules changent. Ici, on en a 2.
On reprends l'exemple du docteur mais on a désormais 2 docteurs 
: 
Les cas d'urgence arrivent indépendamment au hasard
• Supposons que les arrivées suivent un processus d'entrée de Poisson (exponentiel
délais entre les arrivées) et que le temps passé avec le médecin urgentiste est
distribution exponentielle

- Taux d'arrivée moyen = 1 patient tous les 1/2 heure soit si n mets à la même unité 2 patients / heure

- Temps de service moyen = 20 minutes pour traiter chaque patient 
µ= 1 patient / 20 minutes donc 3 patients / heure

Le code et les formules ne sont plus les mêmes parce que maintenant, S = 2.

Pour retranscrire les formules en Python, on lit le cours 
http://courses.washington.edu/inde411/QueueingTheoryPart3.pdf

On remarque que pour la formule donnant P0, il va falloir créer une itération pour qu'elle puisse s'adapter à d'autres cas ( M/M/4 par exemple)

```python
######################
# Paramètres
######################

# Nombre de serveurs
S = 2

# Taux d'arrivée par heure ( 2 patients par heure )
Lambda = 2

# Taux de service  par heure ( 3 rdv par heure)
mu = 3

import math

######################
# Statistiques
######################

# Intensité de serveur, intensité de trafic ( si P < 1  , alors état stable (steady state OK dans ce cas)): 
P = Lambda / (mu * S)

#########
# probabilité que les deux médecins soient inactifs ?
# probability that both doctors are idle?
#########

# Equation décomposée
k1 = ((Lambda / mu )**0 / math.factorial(0))
k2 = ((Lambda / mu )**1 / math.factorial(1))
k3 = ((Lambda / mu )**2 / math.factorial(2))
k4 = round(1 / (1 - P),2)
P0 = k1 + k2 + k3 * k4 # k3 * k4 est prioritaire, c'est normal
P0 = 1 / P0

# Equation complète
P0 = 1 /  (((Lambda / mu )**0 / math.factorial(0)) + ((Lambda / mu )**1 / math.factorial(1)) + ((Lambda / mu )**2 / math.factorial(2)) * round(1 / (1 - P),2))

#########
# nombre prévu de patients en attente d’un médecin ?
# expected number of patients waiting for a doctor? 
#########
Lq = (P0 * ( Lambda / mu ) ** S ) * P / (math.factorial(S) * (1-P) ** S)

#########
# nombre attendu de patients aux urgences
# expected number of patients in the ER
#########
L = Lq + (Lambda / mu)

#########
# temps prévu aux urgences ?
# expected time in the ER?
#########
W = L / Lambda

#########
# temps d'attente prévu ?
#expected waiting time?
#########
Wq = Lq / Lambda

# Imprimer le reporting
print("Intensité de queue:",round(P,2),"\n")
print("probabilité P(0) que les deux médecins soient inactifs ?",P0 * 100 ,"%")
print("nombre Lq prévu de patients en attente d’un médecin ?",Lq)
print("nombre L attendu de patients aux urgences",L,"personnes")
print("temps W prévu aux urgences ?",W ,"heures")
print("temps Wq d'attente prévu ?",Wq ,"heures")
```
