# Imaginons que les frais de port diminuent d'un pourcentage en fonction du nombre d'objets x 



```python
import matplotlib.pyplot as plt
import math

x_cords = range(0,100)

# les frais de port sont de 5 euros par objet

y_lin = [x * 5 for x in x_cords]

# les frais de port sont de 5 euros par objet moins une réduction de 20% pour x
y_cords = [ 5 * x -  ( 5 * x * 20 / 100) for x in x_cords] 


plt.plot(x_cords, y_lin, "-b", label="Port total pour 5 euros par objets")
plt.plot(x_cords, y_cords, label="réduction de 20% pour x")
plt.xlabel("nombre d'objets")
plt.ylabel("cout du port ")
plt.legend(loc="upper left")
plt.show()
```

# Prix qui diminue en fonction du nombre produit


Un producteur veut réduire son prix de gros.
Le prix normal est de 81 euros, 

Mais le prix de gros est moins cher quand la quantité commandée augmente.


```python
import matplotlib.pyplot as plt
import math

x_cords = range(0,1000)

prix_lineaires = [x * 81 for x in x_cords] 
prix_non_lineaires = [x * ( 81 - x / 20) for x in x_cords] 

plt.plot(x_cords, prix_lineaires, label="Prix linéaires")
plt.plot(x_cords, prix_non_lineaires, label="Réduction du prix de gros non linéaire")

plt.xlabel("nombre d'objets produits")
plt.ylabel("prix")
plt.legend(loc="upper left")
plt.show()
```

# Tracer deux équations avec NUMPY

```python
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend();
```

```python
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, x * ( x/10), '-b', label='x * ( x/10)')
ax.plot(x, x * (x/9), '--r', label='x * (x/9)')
ax.axis('equal')
leg = ax.legend();
```
