<h2>1. Problème de portefeuille (Markowitz)</h2>

```python
from docplex.mp.model import Model

mdl = Model("portfolio")

n = 4
x = mdl.continuous_var_list(n, lb=0, ub=1)

Q = [[0.1, 0.02, 0.01, 0.03],
     [0.02, 0.2, 0.04, 0.01],
     [0.01, 0.04, 0.15, 0.02],
     [0.03, 0.01, 0.02, 0.1]]

returns = [0.12, 0.10, 0.15, 0.09]

mdl.minimize(mdl.sum(x[i]*Q[i][j]*x[j] for i in range(n) for j in range(n)))

mdl.add_constraint(mdl.sum(x) == 1)
mdl.add_constraint(mdl.sum(x[i]*returns[i] for i in range(n)) >= 0.11)

mdl.solve()
solution = mdl.solve()
print (solution)
```
