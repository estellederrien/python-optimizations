<h2>📌 Exemple de suite convergente</h2>

<p>Considérons la suite définie par :</p>

<p><strong>u<sub>n</sub> = 1 / n</strong></p>

<p>Quand n devient de plus en plus grand :</p>
<ul>
    <li>u₁ = 1</li>
    <li>u₂ = 0,5</li>
    <li>u₁₀ = 0,1</li>
    <li>u₁₀₀ = 0,01</li>
</ul>

<p>👉 On voit que les termes se rapprochent de <strong>0</strong>.</p>

<p>✔️ On dit que la suite <strong>converge vers 0</strong>.</p>

<h2>🧠 Intuition</h2>

<p>Plus n → ∞, plus 1/n → 0.</p>


<h2>🐍 Code Python simple</h2>

<p>Voici un petit programme pour illustrer ça :</p>

# Exemple de suite u_n = 1/n



```python
# Exemple de suite u_n = 1/n

def suite(n):
    return 1 / n

# Affichage des 10 premiers termes
for n in range(1, 11):
    print(f"u_{n} = {suite(n)}")
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

n_values = list(range(1, 50))
u_values = [1/n for n in n_values]

plt.plot(n_values, u_values, marker='o')
plt.title("Convergence de la suite u_n = 1/n")
plt.xlabel("n")
plt.ylabel("u_n")
plt.grid()

plt.show()
```

<h2>✅ Résumé</h2>

<ul>
    <li>Une suite converge si ses termes se rapprochent d’une valeur.</li>
    <li>Ici : u<sub>n</sub> = 1/n → 0</li>
</ul>

<p>Si tu veux, je peux aussi te montrer :</p>
<ul>
    <li>un exemple de suite <strong>non convergente</strong></li>
    <li>ou une suite <strong>qui converge vers une autre valeur que 0</strong> 👍</li>
</ul>
