# Classifieur de type arbre de décision

On utilise des questions successives que l'on applique à un dataset afin d'obtenir un résultat statistique.

Ici, on utilise un dataset de patients afin de détecter si un cancer  est malingre ou pas.

Source : O Reilly - Introduction to Machine Learning with python

```python
# Chargement de la lib
from sklearn.model_selection import train_test_split # Fonction qui permet d'automatiquement entrainer 75% du dataset
from sklearn.tree import DecisionTreeClassifier
```

```python
# Chargement du dataset
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
# Visualiser l'arbre décision avec GraphViz
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["malignant", "benign"],
feature_names=cancer.feature_names, impurity=False, filled=True)
```
