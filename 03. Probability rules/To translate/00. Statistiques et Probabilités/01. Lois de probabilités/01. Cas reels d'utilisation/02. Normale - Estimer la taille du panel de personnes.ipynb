# Loi Normale

Description :

La loi normale est une distribution de probabilité en forme de cloche.

Source : https://www.askpython.com/python/normal-distribution

Permet de résoudre :

1. Quelle est la pb que échantillon soies "plus que"
2. Quelle est la pb que échantillon soies "moins que"
3. Quelle est la pb que échantillon soies entre une valeur min et une valeur max


Notre problème:

On a une distribution du poids de notre groupe d'employés entre 0 et 100kgs.

```python
# On importe les libs
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python

 # On crée la distribution
data = np.arange(1,100,0.3)
pdf = norm.pdf(data , loc = 70 , scale = 10 )
 
# On dessine la distribution
 
sb.set_style('whitegrid')
sb.lineplot(x=data, y=pdf , color = 'black')
plt.xlabel('Poids')
plt.ylabel('Densité de probabilité')
```

Maintenant, si on nous demandait de choisir une personne au hasard dans cette distribution, alors quelle est la probabilité que le poids de la personne soit inférieure à 60 Kgs ?

```python
MonResultat = norm(loc = 70 , scale = 10 ).cdf(60)
print("Ma probabilité est de : ", MonResultat * 100, "%")
```

Maintenant, si on nous demandait de choisir une personne au hasard dans cette distribution, alors quelle est la probabilité que le poids de la personne soit supérieure à 60 Kgs ?

```python
MonResultat = 1-norm(loc = 70 , scale = 10 ).cdf(60)
print("Ma probabilité est de : ", MonResultat * 100, "%")
```

Maintenant, si on nous demandait de choisir une personne au hasard dans cette distribution, alors quelle est la probabilité que le poids de la personne soit entre 60 et 70 Kgs ?

```python
# On fait un 'RANGE' pour obtenir l'aire sous la courbe (intégrale)
rv = norm(70,10)
resultat = rv.cdf(70) - rv.cdf(60)
print("La probabilité que la personne pèse entre 60 et 70kgs est de : ", resultat * 100 ,"%" )
```
