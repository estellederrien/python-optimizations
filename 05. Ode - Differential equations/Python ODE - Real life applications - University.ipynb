# <b> 1. Second order differential equations</b>
       
       
## Abstract 

Second order ODEs appear in:

- mechanics (Newton's law)

- harmonic oscillator

- RLC circuits

- vibration

- wave equations


Les équations différentielles linéaires du second ordre à coefficients constants sont fondamentales en physique, notamment pour modéliser les systèmes oscillants (ressorts, circuits RLC, amortisseurs de voiture).
Pour illustrer une équation différentielle ordinaire (EDO) à coefficients constants, rien ne vaut le classique système masse-ressort-amortisseur.

C'est l'exemple type car il régit tout : de la suspension de votre voiture aux vibrations d'un gratte-ciel pendant un séisme. Lorsqu'elle est "égale à zéro" (homogène), cela signifie qu'on étudie le mouvement libre du système après une perturbation initiale, sans force extérieure continue.

```python
import numpy as np
import matplotlib.pyplot as plt

# Coefficients de l'équation : y'' - 3y' + 2y = 0
a = 1
b = -3
c = 2

# Calcul du discriminant
delta = b**2 - 4*a*c

# Racines de l'équation caractéristique
r1 = (-b + np.sqrt(delta)) / (2*a)
r2 = (-b - np.sqrt(delta)) / (2*a)

print("Racines :", r1, r2)

# Définition de la solution générale
def y(x, C1, C2):
    return C1*np.exp(r1*x) + C2*np.exp(r2*x)

# Domaine
x = np.linspace(0, 5, 200)

# Exemple avec C1 = 1, C2 = 2
y_values = y(x, 1, 2)

# Graphique
plt.plot(x, y_values)
plt.title("Solution de y'' - 3y' + 2y = 0")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
