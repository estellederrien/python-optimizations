# <div style="background:black;color:white;width:100%;text-align:center"><b> y' = 2x </b></div>


Here, y' is a <b>derivative</b>, 2 is a <b>constant</b>, x is a <b>variable</b>.<br>
"We have a function whose derivative is equal to twice the variable x."<br>
"The unknown is the function ùë¶ whose derivative is equal to 2x."<br>
Link (French) : https://www.youtube.com/watch?v=qHF5kiDFkW8

## <b>Graphical solution</b>
We are looking for the graphical solution of our differential equation. The solution being a function, it draws a function, the function f(x) = x¬≤. It is thanks to these successive derivatives that we reconstruct and draw f(x) = x¬≤ on the interval [0;10].

```python
# -----------------------------------
# Import lib
# -----------------------------------
from scipy import *
from scipy.integrate import odeint # Solve ODE
import matplotlib.pyplot as plt # Plot graphs
import numpy as np
from math import *

# -----------------------------------
# function that returns the derivative for each x supplied to the np.linspace function
# -----------------------------------
def derivative (y,x) :
    return 2 * x

# -----------------------------------
# time points
# -----------------------------------
x = np.linspace ( 0 , 10)

# -----------------------------------
# Initial condition
# -----------------------------------
y0 = 1

# -----------------------------------
# solve ODE
# -----------------------------------
solution = odeint ( derivative , y0 , x )
y = solution [ : , 0 ]

# -----------------------------------
# Plot result
# -----------------------------------
plt.plot ( x , y )
plt.title("ODE y' = 2x - Solution : x¬≤ " )
plt.ylabel("value")
plt.xlabel("time")
plt.show()

# Find the value at time 10 with graphical interpolation We see that this is not the precise result
print(" At time 10, my value will be " ,np.interp(10, x, y))
```

## <b>Symbolic solution</b>

Now, let's find the solution to the differential equation, using mathematical symbols directly.

We're still trying to solve:
$$
y'
= 2x

## Step 1 : Entering ODE into Sympy

This allows Sympy to interpret the diff equation.

```python
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols
from sympy.abc import x
import sympy

x   = sympy.Symbol('x')
y   = sympy.Function('y')

left_ode_side  = sympy.diff(y(x),x)
right_ode_side = 2 * x

edo = sympy.Eq( left_ode_side , right_ode_side )
display(edo)
```

## Step 2 : Computing general solution

We calculate the solution functions of the diff equation.
We find the same solution as Yvan Monka.

```python
solgen = sympy.dsolve(edo,y(x)) 
display(solgen)
```

## Step 3: Computing a particular solution

If we're looking for x = 10, we use the general solution found just above in step 2, and this is also evident on the graph. We then find y = 100.
We can also display x = 3, to see, and we find 9.
We substitute x for the value in the general solution.

```python
part_sol = 10 ** 2
print (" My value after 10 times is " , part_sol)

part_sol = 3 ** 2
print (" My value after 3 times is " , part_sol)
```

## <b>Algebric solution using DSolve</b>

We use dsolve to solve
$$
y'
= 2x
$$

This gives us the general solution. In this example, there is no initial condition; we'll see that later...
We check the results and see that we find the same solution as Yvan Monka, in logical, graphical, symbolic, and algebraic solutions. This proves that the processes and codes are correct.

```python
# -----------------------------------
# Import lib
# -----------------------------------
from sympy import symbols, Function, Eq, dsolve

# -----------------------------------
# Parameters
# -----------------------------------
x = symbols('x')
y = Function('y')

# -----------------------------------
# Create ODE
# -----------------------------------
# Ode left part, then ode right part
diff_equation = Eq(Derivative(y(x), x) , 2*x)

# -----------------------------------
# Solve ODE
# -----------------------------------
# We solve ODE, and get a general solution
solution = dsolve(diff_equation)

# -----------------------------------
# Display solution
# -----------------------------------
print(solution)

# Simplified solution
print(solution.rhs)
```

## <b>Abstract </b>
It is logical to assume the rate of change of population P with respect to time t is proportional to the
current population size P . 

$$ 
\frac{dP}{dt} = rP 
$$ 

 where r is the constant relative growth rate. 

 We usually add the starting population information , and the time range we want to analyze : 

$$ 
P(0) = 100 , t ‚àà [0, 5] 
$$

This IVP results is the exponential relation for population P
as a function of time t.

thus


$$ 
P(t) = P(0)e^{rt}  
$$ 




Link : https://www2.math.upenn.edu/ugrad/calc/m114/Stewart104.pdf

https://www3.nd.edu/~apilking/Calculus2ResourcesOriginal/Calculus2ResourcesOriginal/Lecture%205/NotesL5.pdf

## <b>Instance </b>

Suppose that a population of bacteria grows naturally. <br>
Suppose that the initial population is 100 bacteria, and, after 1 hour, the population has grown to 120. <br>

- How large will the population be after 2 hours? <br>
- How long will it take for the population to reach 1,000,000 bacteria?<br><br>

Link : https://www.math.uci.edu/~ndonalds/math2a/notes/3-8.pdf

## <b>Manual solution </b>

The instance gave me some informations, that will permit me to find the growth rate that is unknown.
We do not know the constant growth rate, so we need to calculate it manually, after that , we'll get the generic formula, to be able to answer "How large will the population be after 2 hours"?  :

<b>Step 1 :</b> 

P(0) is equal to 100 bacterias, so, substituing inside of $$ P(t) = P(0)e^{rt}   $$ 
We get : 

$$ 
P(t) = P(0) = 100e^{r * t}   
$$ 

<b>Step 2 :</b> To find the r growth rate, let's look for P(1) = 120, r is unknown , let's calculate it!

$$ 
120 = P(1) = 100e^{r * 1} = 100e^{r}  -> e^{r} = \frac{120}{100} ->  e^{r} = \frac{6}{5} = 1.2  -> r = (ln 5/6) 
$$ 

<b>Step 3 </b>
Now that we now the r growth rate, let's substitute inside of the general formula : 

$$ 
P(t) = 100(\frac{6}{5})^t 
$$

It is simplier to write than : 

$$ 
P(t) = 100e^{(ln \frac{6}{5})t}
$$

Now that we have the generic formulation, we can answer question 2 : How large will the population be after 2 hours?

$$ P(2) = 100(\frac{6}{5})^2 $$

```python
# Pop After 2 hours
import math 
population_after_2_hours = 100 * (6/5) ** 2
print (" My population at time 2 will be " , population_after_2_hours)
```

## <b> Graphical solution </b>

Now, I am using scipy to plot the differential equation, please notice that I add the r growth rate that is ln(6/5), as we computed before.

```python
# -----------------------------------
# Import lib
# -----------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# -----------------------------------
# Instance
# -----------------------------------
r = np.log(6/5) # growth rate = ln(6/5) = 0.1823215567939546
p0 = 100        # initial condition  : 100 bacterias at time 0

# -----------------------------------
# function that returns the derivative for each x supplied to the np.linspace function
# -----------------------------------
def derivative (y,x) :
    return  y * r

# -----------------------------------
# time points
# -----------------------------------
x = np.linspace ( 0 , 5) # range of 0 to 5

# -----------------------------------
# solve ODE
# -----------------------------------
solution = odeint ( derivative , p0 , x )
y = solution [ : , 0 ]

# -----------------------------------
# plot result
# -----------------------------------
plt.plot ( x , y )
plt.title("Diff equation : y' = rP " )
plt.ylabel("population")
plt.xlabel("time")
plt.show()

```

## <b>Abstract</b>
Here is the differential equation of the Verhulst logistics equation 

 $$ 
 y' = ry (1- \frac{y}{K})   
 $$ 

 $$ 
 t(0) = 100 , t ‚àà [0, 40] 
 $$

Where
- y' which is the evolution of the population according to time t
- y which is the function
- r which is the growth rate
- K The population threshold, or the load capacity, that is why we say "limited resources"
- t (0) which is the population in time 0
- [0.40] which is the number of time steps, in other words the extent of the analysis.

which is generally used to model the self-limited growth of a biological population, that is, limited growth by limited resources.
Initially, growth follows an approximate exponential model with a growth rate r.
As the size of the population is close to the K load capacity K, the population
The curve is flattened, for a model generally used to model the self-limited growth of a biological population.

I can write it like that too : 

 $$ 
 \frac{dy}{dt} = ry (1- \frac{y}{K})   
 $$ 

 $$ 
 t(0) = 100 , t ‚àà [0, 40] 
 $$

 (Wikipedia source : https://fr.wikipedia.org/wiki/Mod%C3%A8le_de_Verhulst):

## <b> Graphical Solution </b>

```python

# -----------------------------------
# Instance
# -----------------------------------
# Growth rate
r = 0.2
# Population threshold
K = 200
# Initial condition
y0 = 100 # At the beggining, there are 100 individuals

#  Time points
x = np.linspace ( 0 , 40) # This is the extent of our analysis

# -----------------------------------
# EDO
# -----------------------------------
# Equation
def deriv (y,x) :
    return  r * y *  ( 1 - (y / K)) # We stupidly copy the differential equation cited above, taken from Wikipedia

# -----------------------------------
# Solve EDO
# -----------------------------------
solution = odeint ( deriv , y0 , x )
y = solution [ : , 0 ]

# -----------------------------------
# Plot solution
# -----------------------------------
plt.plot ( x , y )
plt.title("Individual" )
plt.ylabel("amount")
plt.xlabel("time")
plt.show()
```

## <b> Abstract </b>
Used in biology to model the growth of a population (bacteria, animals) in an environment with limited resources.

The concept: At first, the population grows quickly. But the closer it approaches the maximum capacity of the ecosystem (K), the more growth slows down.

## <b> Model </b>
 $$ 
 \frac{dP}{dt} = rP (1- \frac{P}{K})   
 $$ 

 Where

- P = Population
- r = Groth rate
- K = Capacity (limited ressources)

Initial condition (Initial population) and time(t) study range.
 $$ 
 t(0) = 10 , t ‚àà [0, 30] 
 $$

 ## <b> Instance </b>

 My initial population is 10 subjects. My reproduction rate is 5%, the max capacity of environment is 1000,
 How will my population evolve during the time (t) ?

 ## <b>Scipy solution </b>

 

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Mod√®le de croissance logistique
def croissance_logistique(P, t, r, K):
    dPdt = r * P * (1 - P / K)
    return dPdt

# Param√®tres
P0 = 10      # Initial population
r = 0.5      # Reproduction rate
K = 1000     # Environment max capacity
t = np.linspace(0, 30, 100)

# R√©solution
solution = odeint(croissance_logistique, P0, t, args=(r, K))

# Affichage
plt.plot(t, solution, color='green', label='Population P(t)')
plt.xlabel('Time')
plt.ylabel('Number of elements')
plt.title('Logistic growth model')
plt.grid(True)
plt.legend()
plt.show()
```

$$
\frac{dT}{dt}

## <b> Graphical solution </b>

## <b>Abstract</b>

In economics, differential equations model various aspects such as investment growth, consumption, and economic cycles. 
     

## <b>Computation</b> 
This equation illustrates exponential growth of investments over time.

$$
\frac{dA}{dt} = rA
$$

Where:

    A(t) = amount at time t (in years)
    A0‚Äã = initial amount
    r = interest rate (as a decimal)


<b>Exact Solution (Analytical)</b><br>

Separate variables:
$$ \frac{1}{A} dA = r \, dt $$

Integrate:
$$ \ln |A| = rt + C $$

Exponentiate:
$$ A(t) = A_0 e^{rt} $$
This is the continuous compounding formula.



## <b>Instance</b> 
Real-life case: Retirement savings account

Scenario

You deposit $10,000 into a retirement account

The account earns 5% annual interest, compounded continuously

You leave the money untouched for 20 years

Model:
$$
\frac{dA}{dt} = 0.05A , A(0) = 10000
$$

Solution : 
$$ A(t) = 10000 e^{0.05t} $$

After 20 years:

$$ A(20) = 10{,}000 e^{0.05 \times 20} = 10{,}000 e^{1} \approx 10{,}000 \times 2.718 \approx \$27{,}180 $$


## <b>Solve using Scipy :</b>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
# Parameters
A0 = 10000 # initial amount ($)
r = 0.05 # interest rate (5%)
t_span = (0, 20) # time interval (years)
# ODE definition
def interest_ode(t, A):
    return r * A
    # Solve the ODE
solution = solve_ivp(
    interest_ode,
    t_span,
    [A0],
    t_eval=np.linspace(0, 20, 200)
    )
# Extract results
t = solution.t
A = solution.y[0]
# Plot
plt.figure()
plt.plot(t, A)
plt.xlabel("Time (years)")
plt.ylabel("Amount ($)")
plt.title("Continuous Compound Interest (SciPy)")
plt.show()
```

## <b>Abstract </b>
Gemini AI quotes :
To understand how medicine behaves in the bloodstream, we use a Pharmacokinetic (PK) model. The simplest and most common version is the One-Compartment Model, which treats the entire body as a single volume where the drug is distributed instantly and then eliminated.

Assumptions

- The drug is injected instantly into the bloodstream

- The body removes the drug at a rate proportional to its current concentration

- No additional doses are given

## <b>Math model </b>
The rate of change of the drug concentration in the blood depends on how much is being added (input) and how much is being cleared by the kidneys or liver (output).

The differential equation for a single intravenous (IV) bolus dose is:


$$
\frac{dC}{dt} = -k.C(t)
$$

Where:

    C(t) = C(t) is the concentration of the drug at time t.
    k‚Äã = k is the elimination rate constant (units of 1/time).

Initial condition :

 $$
 C(0) = C_{0}
$$

## <b> Instance </b>

- Initial concentration: 
$$ C_{0} = 50 mg/L $$

- Elimination rate: 
$$ k=0.3‚Äâ $$

- Simulate for 24 hours


```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parameters
k = 0.3          # elimination rate (1/hour)
C0 = [50]        # initial concentration (mg/L)
t_span = (0, 24) # time range (hours)
t_eval = np.linspace(0, 24, 200)

# ODE function
def drug_concentration(t, C):
    return -k * C

# Solve ODE
solution = solve_ivp(
    drug_concentration,
    t_span,
    C0,
    t_eval=t_eval
)

# Plot
plt.plot(solution.t, solution.y[0])
plt.xlabel("Time (hours)")
plt.ylabel("Drug concentration (mg/L)")
plt.title("Drug concentration in blood over time")
plt.grid()
plt.show()
```

## <b> Abstract </b>

In physics and engineering, we use First-Order Ordinary Differential Equations (ODEs) to describe how a system changes over time. When we talk about velocity and acceleration, we are essentially looking at the relationship between a function and its derivative.

## <b>Math model </b>

In calculus, acceleration (a) is the rate of change of velocity (v) with respect to time (t). Mathematically, this is expressed as:

$$
\frac{dv}{dt}
= a(t)
$$

This is a first-order ODE because it involves the first derivative of velocity. If we know the acceleration acting on an object, we can solve this equation to find the velocity at any given time.

## <b>Instance</b>
Real-Life Example: Car Accelerating

Scenario:
A car starts from rest and accelerates along a straight road. Its acceleration depends on time (or it could depend on velocity, but let's keep it simple).

Let‚Äôs define:

- v(t) = velocity of the car at time t 
- a(t) = acceleration of the car at time t

From physics:

$$
\frac{dv}{dt}
= a(t)
$$

<b>Example 1: Constant Acceleration</b>

Say the car accelerates at a constant rate of 2 m/s¬≤:

$$
\frac{dv}{dt}
= 2
$$

Initial condition: 
v(0)=0 m/s

We can solve this ODE analytically: 
v(t)=2t

## <b> Scipy solution</b>

Python Code (Solving ODE Numerically)

We can use scipy.integrate.solve_ivp to solve it.

```python
# Define acceleration function
def acceleration(t, v):
    return 2  # constant acceleration 2 m/s¬≤

# Time span
t_span = (0, 10)  # 0 to 10 seconds
v0 = [0]  # initial velocity

# Solve ODE
sol = solve_ivp(acceleration, t_span, v0, t_eval=np.linspace(0, 10, 100))

# Plot
plt.plot(sol.t, sol.y[0], label='Velocity v(t)')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Car accelerating with constant acceleration')
plt.grid(True)
plt.legend()
plt.show()
```

<b>Example 2: Velocity-Dependent Acceleration</b>

Sometimes acceleration depends on velocity, like air resistance:

$$
\frac{dv}{dt}
= 5-0.1v
$$
	‚Äã

Here, the car accelerates with a maximum thrust of 5 m/s¬≤ but is slowed down by air resistance proportional to velocity.

Initial velocity: 
v(0)=0

## <b> Scipy solution</b>
Here, the velocity asymptotically approaches 50 m/s, where acceleration becomes zero.

```python
def acceleration_velocity(t, v):
    return 5 - 0.1 * v

v0 = [0]
sol = solve_ivp(acceleration_velocity, (0, 50), v0, t_eval=np.linspace(0, 50, 200))

plt.plot(sol.t, sol.y[0], label='Velocity v(t)')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity-dependent acceleration (air resistance)')
plt.grid(True)
plt.legend()
plt.show()
```

Great! Let‚Äôs make a motion-style diagram where we show velocity and acceleration vectors at several time points, which is very intuitive for real-life visualization.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Acceleration function (depends on velocity)
def acceleration_velocity(t, v):
    return 5 - 0.1 * v

# Initial velocity
v0 = [0]

# Time points
t_eval = np.linspace(0, 50, 200)

# Solve ODE
sol = solve_ivp(acceleration_velocity, (0, 50), v0, t_eval=t_eval)

# Compute acceleration at each time point
a_values = 5 - 0.1 * sol.y[0]

# Choose sample points for arrows
sample_indices = np.arange(0, len(t_eval), 20)  # every 20 points

# Plot velocity
plt.figure(figsize=(12,6))
plt.plot(sol.t, sol.y[0], label='Velocity v(t) [m/s]', color='blue')

# Plot acceleration arrows
for i in sample_indices:
    plt.arrow(sol.t[i], sol.y[0][i], 0, a_values[i]*1.5, 
              head_width=0.8, head_length=1.0, fc='red', ec='red', alpha=0.7)

plt.xlabel('Time (s)')
plt.ylabel('Velocity / Acceleration')
plt.title('Velocity and Acceleration Vectors over Time')
plt.grid(True)
plt.legend(['Velocity v(t)', 'Acceleration a(t) (red arrows)'])
plt.show()
```

How this diagram works:

Blue curve ‚Üí velocity over time.

Red arrows ‚Üí acceleration at specific time points, scaled to visualize magnitude.

The arrow length shows how strong the acceleration is: big arrows at start (high acceleration) and small arrows as velocity approaches terminal velocity.
