## <b>Solution Logique.</b>

Gr√¢ce au tableau des d√©riv√©es usuelles du Lyc√©e, on sait que la fonction x¬≤ a pour d√©riv√©e 2x.<br>
Par d√©duction, la fonction qui est solution de l'√©quation diff√©rentielle est donc x¬≤

## <b>Solution Graphique </b>

On cherche la solution graphique de notre √©quation diff√©rentielle. La solution √©tant une fonction ,√ßa dessine une fonction, la fonction f(x) = x¬≤ . C'est gr√¢ce √† ces d√©riv√©es successives que l'on reconstitue et dessine f(x) = x¬≤ sur l'intervalle [0;10].

```python
# Import des libs
from scipy import *
from scipy.integrate import odeint #Pour r√©soudre l‚Äô√©quation diff√©rentielle
import matplotlib.pyplot as plt # Permet de tracer des graphes
import numpy as np
from math import *
```

```python
# fonction qui retourne la d√©riv√©e pour chaque x fourni √† la fonction np.linspace
def derivee (y,x) :
    return 2 * x

#  time points
x = np.linspace ( 0 , 10)

# Condition initiale
y0 = 1

# solve ODE
solution = odeint ( derivee , y0 , x )
y = solution [ : , 0 ]

# plot results
plt.plot ( x , y )
plt.title("Equation diff y' = 2x - Solution : x¬≤ " )
plt.ylabel("valeur")
plt.xlabel("temps")
plt.show()

# Trouver la valeur au temps 10 avec l'interpolation graphique On voit que ce n'est pas le r√©sultat pr√©cis
print(" Au temps 10, ma valeur sera de " ,np.interp(10, x, y))
```

## Etape 1 : Entrer l'eq diff dans Sympy

Ca permet √† Sympy d'interpr√©ter l'equation diff.

```python
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols
from sympy.abc import x
import sympy

x   = sympy.Symbol('x')
y   = sympy.Function('y')

partie_gauche  = sympy.diff(y(x),x)
partie_droite = 2 * x

edo = sympy.Eq( partie_gauche , partie_droite )
display(edo)
```

On calcule les fonctions solution de l'eq diff.

```python
solgen = sympy.dsolve(edo,y(x)) 
display(solgen)
```

## <b>Solution alg√©brique avec DSolve</b>

On se sert de dsolve pour r√©soudre 
$$
y'

$$

Cela nous donne la solution g√©n√©rale, dans cet exemple, il n'y a pas de condition initiale, on va voir √ßa apr√®s...

```python
from sympy import symbols, Function, Eq, dsolve

# On param√™tre les symboles et fonction
x = symbols('x')
y = Function('y')

# On entre l'√©quation diff√©rentielle dans SYMPY
# Eq(PARTIE GAUCHE DE L'EQUATION DIFFERENTIELLE , PARTIE DROITE DE L'EQUATION DIFFERENTIELLE)
equation_differentielle = Eq(Derivative(y(x), x) , 2*x)

# On r√©souds l'√©quation diff√©rentielle, √ßa nous donne une solution g√©n√©rale
result = dsolve(equation_differentielle)

# affichage classique
print(result)

# affichage simplifi√©
print(result.rhs)
```

```python
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols
from sympy.abc import x
import sympy

x   = sympy.Symbol('x')
y   = sympy.Function('y')

left  = sympy.diff(y(x),x)

right = -2 * y(x) / 5 # La fonction y s'√©crit y(x)

edo = sympy.Eq( left , right )
display(edo)
```

## Etape 2 : Calcul de la Solution g√©n√©rale

## Etape 1 :  Entrer la condition

La condition est une constante.

```python
x0 = 1
y1 = 3
consts = sympy.solve( sympy.Eq( y1, solgen.rhs.subs(x,x0)) , dict=True)[0]
consts
```

## Etape 2 : Calculer La solution unique avec la constante

On prends en compte la constante d√©finie pr√©c√©demment.
On voit qu'on trouve la m√™me solution que dans l'exercice .

```python
solpar = solgen.subs(consts)

display("Unique solution telle que ùëì(1) = 3")
solpar
```

## <strong>L'√©quation diff√©rentielle <span style="color:red"> 3ùë¶' + 5ùë¶ = 0<span></strong>
-------------------------------------------------------------------------------------------------------

On consid√®re l‚Äô√©quation diff√©rentielle 3ùë¶' + 5ùë¶ = 0.

- 1) D√©terminer la forme g√©n√©rale des fonctions solutions de l‚Äô√©quation.

- 2) D√©terminer l‚Äôunique solution ùëì telle que ùëì(1) = 2.

Lien : https://www.maths-et-tiques.fr/telech/20Prim-EdTC.pdf

On retrouve le m√™me r√©sultat !


```python
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols
from sympy.abc import x
import sympy

x   = sympy.Symbol('x')
y   = sympy.Function('y')

left  = sympy.diff(y(x),x)
right = -5 * y(x) / 3 # La fonction y s'√©crit y(x)

edo = sympy.Eq( left , right )
display(" Affichage de l'eq diff",edo)

solgen = sympy.dsolve(edo,y(x)) 
display(" Solution g√©n√©rale",solgen)

x0 = 1
y1 = 2
consts = sympy.solve( sympy.Eq( y1, solgen.rhs.subs(x,x0)) , dict=True)[0]
consts
solpar = solgen.subs(consts)
display("Unique solution telle que ùëì(1) = 2")
solpar

```

## R√©soudre graphiquement

```python

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 
  
def returns_dydt(y,t): 
    dydt = -y * t + 13
    return dydt 
  
# initial condition 
y0 = 1
  
# values of time 
t = np.linspace(0,5) 
  
# solving ODE 
y = odeint(returns_dydt, y0, t) 
  
# plot results 
plt.plot(t,y) 
plt.xlabel("Time") 
plt.ylabel("Y") 
plt.show()
```

## Exemple 2.2 : Solution chimique  et temps exact avant de trouver un r√©sultat pr√©cis

Trouver au bout de combien de temps la valeur de y se r√©duira √† 1

https://kitchingroup.cheme.cmu.edu/blog/2011/08/31/Solving-an-ode-for-a-specific-solution-value/


La solution analytique d'une ODE est une fonction qui peut √™tre r√©solue pour obtenir une valeur particuli√®re, par ex. si la solution d'une ODE est y(x) = exp(x), vous pouvez r√©soudre la solution pour trouver la valeur de x qui fait y(x)=2

. Dans une solution num√©rique d'une ODE, nous obtenons un vecteur de valeurs de variables ind√©pendantes et les valeurs de fonction correspondantes √† ces valeurs. Pour r√©soudre une valeur de fonction particuli√®re, nous avons besoin d‚Äôune approche diff√©rente. Cet article montrera une fa√ßon de le faire en python.

√âtant donn√© que la concentration d'une esp√®ce A dans un volume constant, le r√©acteur discontinu ob√©it √† cette √©quation diff√©rentielle $$\frac{dC_A}{dt}=- k C_A^2$$ avec la condition initiale CA(t=0)=2,3 mol/L et k=0,23 L/mol/s, calculez le temps qu'il faut pour CA
√† r√©duire √† 1 mol/L.

Nous obtiendrons une solution, puis cr√©erons une fonction d'interpolation et utiliserons fsolve pour obtenir la r√©ponse.

```python
from scipy.integrate import odeint
from scipy.interpolate import interp1d
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt

k = 0.23
Ca0 = 2.3

def dCadt(Ca, t):
    return -k * Ca**2

tspan = np.linspace(0, 10)

sol = odeint(dCadt, Ca0, tspan)
Ca = sol[:,0]

plt.plot(tspan, Ca)
plt.xlabel('Time (s)')
plt.ylabel('$C_A$ (mol/L)')


ca_func = interp1d(tspan, Ca, 'cubic')

itime = np.linspace(0, 10, 200)

plt.figure()
plt.plot(tspan, Ca, '.')
plt.plot(itime, ca_func(itime), 'b-')

plt.xlabel('Time (s)')
plt.ylabel('$C_A$ (mol/L)')
plt.legend(['solution','interpolated'])

plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# function that returns dy/dt
def model(y,t,k):
    dydt = -k * y
    return dydt
# initial condition
y0 = 5
# time points
t = np.linspace(0,20)
# solve ODEs
k = 0.1
y1 = odeint(model,y0,t,args=(k,))
k = 0.2
y2 = odeint(model,y0,t,args=(k,))
k = 0.5
y3 = odeint(model,y0,t,args=(k,))
# plot results
plt.plot(t,y1,'r-',linewidth=2,label='k=0.1')
plt.plot(t,y2,'b--',linewidth=2,label='k=0.2')
plt.plot(t,y3,'g:',linewidth=2,label='k=0.5')
plt.xlabel('time')
plt.ylabel('y(t)')
```

## Solution graphique

La solution d'une eq diff √©tant une fonction, on dessine une fonction...

On prends garde √† bien respecter la nomenclature des noms des variables et des fonctions, pour que cela corresponde au mod√®le math√©matique, comme en recherche op√©rationnelle.

```python
# Taux de croissance
r = 0.2
# Seuil de population
K = 200

# Equation
def deriv (y,x) :
    return y * r * ( 1 - (y / K)) # On recopie b√™tement l'√©quation diff√©rentielle cit√©e plus haut, tir√©e de Wikip√©dia

#  Time points
x = np.linspace ( 0 , 40) # C'est l'√©tendue de notre analyse

# Condition initiale
y0 = 100 # Au d√©but, il y a 100 individus

# R√©soudre l'√©quation diff
solution = odeint ( deriv , y0 , x )
y = solution [ : , 0 ]

# Dessiner le r√©sultat
plt.plot ( x , y )
plt.title("Individus" )
plt.ylabel("nombre")
plt.xlabel("temps")
plt.show()
```

## <b>Cas Pratique 3 : Loi de R√©chauffement/Refroidissement de Newton</b>

Le r√©chauffement ou le refroidissement d'un objet se fait plus ou moins vite en fonction du temps et de la temp√©rature ambiante.
La loi de refroidissement de Newton stipule que le taux d‚Äô√©volution de la temp√©rature d‚Äôun corps est
proportionnel √† la diff√©rence entre la temp√©rature de ce corps et celle du milieu environnant.

Version de Michel Van Biezen ( Refroidissement ) :
$$
\frac{dT}{dt}
= -k(T(t) - T(b))
$$

o√π
- T est la temp√©rature
- dt est le temps
- dT / dt est le changement de la temp√©rature en fonction du temps ( respect to time in english)
- k est une constante ( taux (rate))
- T(t) est la temp√©rature au temps t et T(b) la temp√©rature de l'environnement

## Solution Graphique


```python

K = 0.044629 # Taux de refroidissement qu'on ne peut trouver qu'apr√®s un calcul math√©matique - voir partie apr√®s
T = 5 # Temp√©rature de l'objet
A = 30 # T√©mp√©rature ambiante

# Equation
def deriv (y,x) :
    return - K * ( T - A )

#  Time points
x = np.linspace ( 0 , 60)

# Condition initiale
y0 = T

# R√©soudre l'√©quation diff
solution = odeint ( deriv , T , x )
y = solution [ : , 0 ]

# Dessiner le r√©sultat
plt.plot ( x , y )
plt.title("Temp√©rature" )
plt.ylabel("valeur")
plt.xlabel("temps")
plt.show()
```

 ## Solution Num√©rique

On remplace les valeurs lues dans l'√©nonc√© du probl√®me du th√© dans l'eq diff de NewTon, √ßa donne

$$
\frac{dT}{dt}
= K(T - 30)
$$

Avec K qui est une constante qu'on ne connait pas encore mais qu'on va calculer ensuite.
Le probl√®me, est qu'on ne connait pas cette valeur K, du coup, on ne connait pas l'√©quation 
de la temp√©rature en fonction du temps, mais comme on a des informations dans l'√©nonc√©, on va pouvoir

On utilise d'abord leur corollaire :
<div style="text-align:center">

</div>

On trouve T(t)
\begin{gather*}
T(t) = [T(0)-30]\,e^{Kt} + 30
=30-25 e^{Kt}
\end{gather*}


Ensuite, on cherche K, pour cela on se sert des donn√©es de l'√©nonc√© qui disent que t(5) = 10 ( Ce qui veut dire qu'au temps 5, la temp√©rature du th√© est de 10 degr√©s), On note qu'Il s'agit d'un cas de <b>r√©chauffement</b>, pas de <b>refroidissement</b>.

On a donc une equation a une inconnue √† r√©soudre :

\begin{gather*}
T(5) = 30-25 e^{K * 5} = 10
\end{gather*}

```python
# T5 = 30 - 25 ‚Öá  ( 5 * K ) = 10

# D√©roul√© du calcul √† la main
# 25 ‚Öá  ( 5 * K ) = 10 - 30
# 25 ‚Öá  ( 5 * K ) = -20
# ‚Öá  ( 5 * K ) = 20 / 25
# 5 * K = log(20/25)
# K = 1/5log * 4/5 
# K = - 0.044629

# Le but va √™tre de r√©aliser ce calcul pr√©c√©dent avec Python et/ou directement SYMPY (A venir)
```

## Etape deux : trouver la constante K .

\begin{align*}
25.3=T(2)&= 20+7 e^{2K} & \text{rearrange}\\

# <b>Cas pratique 2 : Charge d'un condensateur.</b>
------------------------------------------------------------------------------------------------------------------------

<div style="text-align:center">
<img src="img/condensateur.jpg">
</div>

On r√©sume la charge d'un condensateur en fonction du temps par l'equation diff suivante :


$$
\frac{du(t)}{dt} = \frac{U - u(t)}{R\,C}
$$

avec pour condition initiale

$$
u(t=0)=0
$$

car le condensateur est d√©charg√© au temps 0

O√ª 
- du(t)/dt signifie "l'√©volution de la tension en fonction du temps"
- dt signifie " en fonction du temps"
- u signifie la tension exprim√©e en volts
- R est la resistance en Ohms
- C est li√© √† l'intensit√© en Farads( Lire la doc)

Source : 
https://cpge.frama.io/fiches-cpge/Python/%C3%89quation%20diff%C3%A9rentielle/0-Equation%20diff%C3%A9rentielle/#exemple-systeme-masseressort



## Solution graphique



```python
from scipy.integrate import solve_ivp
import numpy as np

# Constantes du syst√®me
R = 10  # Ohms
C = 0.01  # F
U = 5  # V

# D√©finition de l'√©quation diff√©rentielle
def equation(t, u):
    return (U-u)/(R*C) # On reconnait bien l'√©quation diff√©rentielle

t0 = 0
tf = 0.8  # seconde
u0 = 0  # Condition initiale

# R√©solution
tension = solve_ivp(equation, [t0, tf], [u0])

# Calcul du courant traversant le condensateur
courant = (U-tension.y[0])/R

import matplotlib.pyplot as plt

# Trac√© de la tension aux bornes du condensateur
plt.plot(tension.t, tension.y[0], label="u(t)")
plt.plot(tension.t, courant, label="i(t)")
plt.ylabel('u(t) en V et i(t) en A')
plt.xlabel("Temps (s)")
plt.title("Tension aux bornes du condensateur et courant le traversant")
plt.grid(which="both")
plt.legend()
plt.show()
```

# L'√©quation diff ne donne pas le m√™me r√©sultat

On voit sur des forums que l'√©quation diff√©rentielle y' = 0.02x serait pratique lors du calcul de croissance de capital d' un taux de 2%, et on sous entends, que le r√©sultat serait le m√™me qu'une suite g√©om√©trique, c'est ce qu'on va voir en dessous.

On voit que le r√©sultat n'est absolument pas le m√™me, ce qui est logique puisque les calculs ne sont pas les m√™mes.


$$ y' = 0.02x $$


```python
# Notre taux de 2 %
taux = 0.02

# fonction qui retourne la d√©riv√©e pour chaque x fourni √† la fonction np.linspace
def derivee (y,x) :
    return taux * x

#  time points
x = np.linspace ( 0 , 5)

# Condition initiale
y0 = 1

# solve ODE
solution = odeint ( derivee , y0 , x )
y = solution [ : , 0 ]

# plot results
plt.plot ( x , y )
plt.title("Equation diff y' = 0.02x  " )
plt.ylabel("valeur")
plt.xlabel("temps")
plt.show()

# Trouver la valeur au temps 3 avec l'interpolation graphique
# On voit que ce n'est pas le r√©sultat pr√©cis
print(" Au temps 3, ma valeur sera de " ,np.interp(3, x, y))
```
